## Asking and Answering Questions

Here are some interesting questions about the dataset



```python
Question 1: Weekdays having more cases
```

```python
Week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
plt.figure(figsize=(12, 6))
plt.title("Weekday having most number of new cases")

data = raw_data.groupby('weekday')['New cases'].sum()
pal = sns.color_palette("Purples_d", len(data))
rank = data.argsort().argsort()
g = sns.barplot('weekday', 'New cases', palette=np.array(pal[::-1])[rank], data=raw_data)
g.set_xticklabels(Week);
```

```python
Question 2: Steepest increase in cases
```

```python
state.index = state.index.astype(str)
```

```python
steep_dict = {}
for i in trends:
    steep_dict[i] = (state.loc['2020-08-06'][i] - state.loc['2020-07-07'][i])/30
```

```python
steep_statedf = pd.DataFrame(steep_dict, index =['slope']).T
```

```python
steep_statedf.sort_values(by = 'slope', ascending=False)
```

```python
plt.figure(figsize=(12, 6))

for i in steep_statedf.index[:8]:
    plt.plot(state.index, state[i])

plt.xlabel('weekday')
plt.ylabel('Total Confirmed cases till  now')

plt.title('Trend of states')
plt.legend(trends);
```
